# Analyzing Students Mental Health in SQL


The study found that international students have a higher risk of mental health difficulties than the general population, and that social connectedness (belonging to a social group) and acculturative stress (stress associated with joining a new culture) are predictive of depression.

# IMPORTING LIBRARY 

```python
import pandas as pd
import sqlite3
```

```python
df=pd.read_csv('students.csv')
```

# CONNECTING SQLITE

```python
cnn = sqlite3.connect('jupyter_sql_tutorial.db')

df.to_sql('students', cnn)
%load_ext sql

%sql sqlite:///jupyter_sql_tutorial.db

```

# SQL ANALYSIS 

```python
%%sql

SELECT * 
FROM students
LIMIT 10;
```

```python
%%sql

SELECT  COUNT(*) AS total_records
FROM students;
```

```python
%%sql


SELECT  inter_dom , COUNT(*) AS count_inter_dom
FROM students
GROUP BY inter_dom;
```

```python
%%sql


SELECT inter_dom ,MIN(todep) AS min_phq , MAX(todep) AS max_phq , 
 ROUND(AVG(todep) ,2) AS avg_phq
FROM students
GROUP BY inter_dom;
```

```python
%%sql


SELECT inter_dom ,MIN(tosc) AS min_sc , MAX(tosc) AS max_sc , 
ROUND(AVG(tosc) ,2) AS avg_sc
FROM students
GROUP BY inter_dom;
```

```python
%%sql


SELECT inter_dom ,MIN(toas) AS min_as , MAX(toas) AS max_as , ROUND(AVG(toas) ,2)
AS avg_as
FROM students
GROUP BY inter_dom;
```

```python
%%sql


SELECT *
FROM students
WHERE inter_dom= 'Inter'
LIMIT 10;
```

```python
%%sql


SELECT inter_dom , suicide , COUNT(suicide)AS count_suicide
FROM students
GROUP BY inter_dom , suicide ;
```

```python
%%sql


SELECT inter_dom ,  AVG(age) AS avg_age
FROM students
GROUP BY inter_dom ;
```

```python
%%sql


SELECT stay, 
       ROUND(AVG(todep), 2) AS average_phq, 
       ROUND(AVG(tosc), 2) AS average_scs, 
       ROUND(AVG(toas), 2) AS average_as
FROM students
WHERE inter_dom = 'Inter'
GROUP BY stay
ORDER BY stay DESC;
```
